# 1. Problem Statemnt

- Header Cleaning
- Converting uppercase into lowercase 
- Removing blank spaces and replacing it with _ underscore

```python
#===========================================================
# STEP 1: HEADER CLEANING
#============================================================

print(df.columns.tolist())
```

```python

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("FIXED APPLIED")
print(df.columns.tolist())
```

# 2. Problem Statement
- Type conversion and currency cleaning

```python
#==========================================================
# STEP 2: TYPE CONVERSION AND CURRENCY CLEANING
#==========================================================

dirty_spend_mask = df["spend"].astype(str).str.contains(r"\$")
print(df.loc[dirty_spend_mask,["campaign_id","spend" ]].head(5))
```

```python
df["spend"] = df["spend"].astype(str).str.replace(r'[^\d.-]','', regex= True)
df["spend"] = pd.to_numeric(df["spend"])

print("FIX APPLIED")
print(df.loc[dirty_spend_mask,["campaign_id","spend" ]].head(5))
```

# 3. Problem Statement
Correcting the different names with the catagagori wise and putting some logics.

```python


print(df["channel"].unique())

cleanup_map = {
     "Facebook": "Facebook",
     "Insta_gram": "Instagram",
     "Gogle": "Google Ads",
     "E-mail": "Email",
     "Tik_Tok": "TikTok",
     "N/A": np.nan
}

df["channel"] = df["channel"].replace(cleanup_map)

print("FIXED APPLIED")
print(df["channel"].unique())
```

```python
print(df["active"].unique())
```

```python
#===========================================================
# STEP 4: HANDELING MIXED BOOLEANS
#============================================================

print(df["active"].unique())

bool_map = {
     "yes": True,
     "Y": True,
     "1": True,
     1: True,
     "No": False,
     "0": False,
     0: False
}

df["active"] = df["active"].map(bool_map).fillna(False).astype(bool)

print("FIX APPLIED")
print(df["active"].unique())


```

fixing date and time

```python
#==========================================================
# STEP 5: DATE AND PARSTING
#============================================================

print(df["start_date"].dtype)

df["start_date"] = pd.to_datetime(df["start_date"], errors="coerce")
df["end_date"]   = pd.to_datetime(df["end_date"],dayfirst= True, errors="coerce")


print("FIXED APPLIED")
print(df["start_date"].dtype)
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
#==============================================================
# LOCAL INTEGRITY (CLICKS VS IMPRESSIONS)
#=============================================================

impossible_mask = df["clicks"] > df["impressions"]
print(df.loc[impossible_mask,["campaign_id","impressions","clicks"]].head(5))
```

```python
#=============================================================
# STEP 7: LOGICAL INTEGRITY (TIME TRAVEL)
#===========================================================

time_travel_mask = df["end_date"] < df["start_date"]
print(df.loc[time_travel_mask,["campaign_id","start_date","end_date"]].head(5))
```

```python

df.loc[time_travel_mask, "end_date"] = df.loc[time_travel_mask,"start_date"] + pd.Timedelta(days= 30)
print(df.loc[time_travel_mask,["campaign_id","start_date","end_date"]].head(3))
```

```python
#===========================================================
# STEP 8: HANDELIGN OUTLIIERS (WINSORIZING)
#===========================================================

Q1 = df["spend"].quantile(0.25)
Q3 = df["spend"].quantile(0.75)

IQR = Q3 - Q1
upper_limit = Q3 + (3 * IQR)

outlier_mask = df["spend"] > upper_limit
print(df.loc[outlier_mask,["campaign_id", "spend"]].head(5))

print("FIXED APPLIED")
df.loc[outlier_mask, "spend"] = upper_limit
print(df.loc[outlier_mask,["campaign_id", "spend"]].head(3))
```

```python
#===========================================================
# STEP 9: STRING PARSING (FEATURE EXTRACTION)
#===========================================================

print(df["campaign_name"].head(3))

df["season"] = df["campaign_name"].str.extract(r'Q\d_([^_]+)_')
print("FIX APPLIED")

print(df["campaign_name"].head(3))
```
